```python
retroplambda=np.array([2/466,5/475,38/474])
retroplambda2=np.array([32/466,78/475,85/474])
width=0.3

fig,ax=plt.subplots()
plt.bar(range(2,len(retroplambda)+2),retroplambda,width=width,label="# of generated RO")
plt.bar(range(2,len(retroplambda2)+2),-1*retroplambda2,width=width,label="# of disappeared RO")
plt.xlabel("# of each commits cluster")
for p in ax.patches:
    ax.annotate(s=np.round(p.get_height(),decimals=4),
                xy=(p.get_x()+p.get_width()/2.,p.get_height()),
                   ha='center',
                   va='center',
                   xytext=((-1,1)),
                textcoords='offset points')
plt.ylabel("# of changed RO\n/# of squashed commits" )
plt.title("retroplambda")
plt.legend(loc='best')
```

```python
def drawBar(generated,disappear,name:str):
    width=0.3
    fig,ax=plt.subplots()
    plt.bar(range(2,len(generated)+2),generated,width=width,label="# of generated RO")
    plt.bar(range(2,len(disappear)+2),-1*disappear,width=width,label="# of disappeared RO")
    plt.xlabel("# of each commits cluster")
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    for p in ax.patches:
        ax.annotate(s=np.round(p.get_height(),decimals=4),
                    xy=(p.get_x()+p.get_width()/2.,p.get_height()),
                       ha='center',
                       va='center',
                       xytext=((-1,1)),
                    textcoords='offset points')
    plt.ylabel("# of changed RO\n/# of squashed commits" )
    plt.title(name)
    plt.legend(loc='best')
    plt.savefig(name+".png")
```

```python
a=np.array([2/466,5/475,38/474])
b=np.array([32/466,78/475,85/474])
drawBar(a,b,'retroplambda')
```

```python
a=np.array([0/230,24/235,0/240,7/241])
b=np.array([4/230,5/235,6/240,17/241])
drawBar(a,b,'mbassador')
```
